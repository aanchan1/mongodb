3)      Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

```python
pass_fail_details=records.aggregate(
   [
      {
         "$project":
           {
             "name": True ,
             "Exam":
               {
                 "$cond": [ { "$gte": [ "$Exam", 40 ] }, "Pass", "Fail" ]
               },
            "Quiz":
               {
                 "$cond": [ { "$gte": [ "$Quiz", 40 ] }, "Pass", "Fail" ]
               },
            "Homework":
               {
                 "$cond": [ { "$gte": [ "$Homework", 40 ] }, "Pass", "Fail" ]
               }
           }
      }
   ]
)
for k in pass_fail_details:
  print(k)
```

4)       Find the total and average of the exam, quiz and homework and store them in a separate collection.

```python
total_avg=records.aggregate(
   [
      {
         "$project":
           {
             "name": True ,
             'total' : { '$add' : [ '$Exam', '$Quiz',"$Homework" ]
              },
            'average': { '$avg' : [ '$Exam', '$Quiz',"$Homework" ]
              }
           }
      }
   ]
)
for k in total_avg:
  print(k)
```
